```python
df1 = pd.DataFrame(dict_indices.items(), columns=["Simbolo", "índice"])
df2 = pd.DataFrame(dict_descricao.items(), columns=["Simbolo", "Descrição"])
df_indice_descricao = pd.merge(df1, df2, on="Simbolo")
df_indice_descricao = df_indice_descricao.set_index('Simbolo')
```

```python
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
url = "https://sistemaswebb3-listados.b3.com.br/indexPage/stocks"
driver.get(url)
```

```python
time.sleep(5)
itens_por_paginas = driver.find_element(By.XPATH, '/html/body/app-root/app-stocks-index/div/form/div[4]/div[1]/div/div[1]/select').click()
botao_120 = driver.find_element(By.XPATH, '/html/body/app-root/app-stocks-index/div/form/div[4]/div[1]/div/div[1]/select/option[4]').click()
```

```python
numero_paginas = driver.find_element(By.XPATH, '/html/body/app-root/app-stocks-index/div/form/div[4]/div[2]/div/pagination-controls/pagination-template/ul/li[7]/a/span[2]')
numero_paginas = int(numero_paginas.text)
numero_paginas
```

```python
lista_de_tabela_por_pagina = []
for pagina in range(0, numero_paginas):
    time.sleep(5)
    tabela = driver.find_element(By.XPATH, '/html/body/app-root/app-stocks-index/div/form/div[3]/div/table')

    html_tabela = tabela.get_attribute("outerHTML")

    tabela_final = pd.read_html(html_tabela)[0]

    lista_de_tabela_por_pagina.append(tabela_final)
    
    botao_avancar_pagina = driver.find_element(By.XPATH, "/html/body/app-root/app-stocks-index/div/form/div[4]/div[2]/div/pagination-controls/pagination-template/ul/li[8]").click()
    
base_de_dados_completa = pd.concat(lista_de_tabela_por_pagina)

base_de_dados_completa
```

```python
print('Temos os seguintes índices no site da B3')
display(df_indice_descricao)
print('Qual índice deseja consultar? ')
print('Exemplo Índice Bovespa, digite:')
print('IBOV')
time.sleep(2)
indice_consulta = str(input()).upper()
if indice_consulta in df_indice_descricao.index:
    indice = base_de_dados_completa[base_de_dados_completa['Índices dos quais participa'].str.contains(indice_consulta, na = False)]
    print(f'O índice consultado possui {len(indice)} ações')
    print(f'Esse indíce é formado pela regra: {df_indice_descricao.loc[indice_consulta][1]}')
    indice.index = range(0, len(indice), 1)
    print(f'As empresas que formam o índice {indice_consulta} são: ')
    with pd.option_context('display.max_rows', None,'display.max_columns', None):
        display(indice[['Empresa','Código']])
else:
    print('Índice não encontrado')
```
